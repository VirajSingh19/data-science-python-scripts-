# Train / Test data for Suprvised learning

some  fake data set :

```python
%matplotlib inline
import numpy as np
from pylab import *

np.random.seed(2)

pageSpeeds = np.random.normal(3.0, 2.0, 100)
purchaseAmount = np.random.normal(50.0, 30.0, 100) / pageSpeeds


scatter(pageSpeeds, purchaseAmount)
```

Splitting the data into trainig and testing data

```python
trainX = pageSpeeds[:80]
testX = pageSpeeds[80:]

trainY = purchaseAmount[:80]
testY = purchaseAmount[80:]
```

Here's our training dataset:

```python
scatter(trainX, trainY)
```

And our test dataset:

```python
scatter(testX, testY)
```

Now we'll try to fit an 8th-degree polynomial to this data (which is almost certainly overfitting, given what we know about how it was generated!)

```python

x = np.array(trainX)
y = np.array(trainY)

p4 = np.poly1d(np.polyfit(trainX, trainY, 8))
print(p4)
```

Let's plot our polynomial against the training data:

```python
import matplotlib.pyplot as plt

xp = np.linspace(0, 7,100)
axes = plt.axes()
axes.set_xlim([0,7])
axes.set_ylim([0, 200])

plt.scatter(x, y)
plt.plot(xp, p4(xp), c='r')
plt.show()
```

And against our test data:

```python
testx = np.array(testX)
testy = np.array(testY)

axes = plt.axes()
axes.set_xlim([0,7])
axes.set_ylim([0, 200])
plt.scatter(testx, testy)
plt.plot(xp, p4(xp), c='r')
plt.show()
```

```python
from sklearn.metrics import r2_score

r2 = r2_score(testy, p4(testx))

print(r2)
```

```python
from sklearn.metrics import r2_score

r2 = r2_score((trainY), p4((trainX)))

print(r2)
```
